Lets start working on contracts
=> Two Folders Requests and Responses
   => Naming conventions XXXXRequest XXXXResponse
   => use properties with required and {get; init;} => does not allow changes during runtime
   => use one class for Create, Update => Better option to allow scalability, do not use the same contract
   => use IEnumerable is favored
   => Requests and responses can have similar structures
   => Collections should be defined with a property inside called items

Create in memory database
=> Repository Pattern
=> class Repositories

Place interfaces inside and Implementation

=> Create models folder => this can have {get; set;} => List etc.
=> Create the IRepostiory => where create , getbyId, getall, update, delete
=> this then can be implemented in a class => start with a memory
   => private readonly List with the implementation itself
=> Use tasks to allow future implementations which most database use
=> This will be injected and needs to be registered in the Program
   builder.Services.AddSingleton<IInterface, IInterface> => Not a good way
   create an extensions ApplicationServiceCollectionExtensions => AddApplication
   => then add the services
   => this is done in order for the API to not know about the application which is a good practice

Now lets advance to the controller
=> Create one Controller => Attribute [ApiController] and base ControllerBase
=> inject the service itself
=> Start Creating Route to introduce api
=> use Task<IActionResult> as preferable for Post
=> use attribute [FromBody] CreateMovieRequest request
=> use attribute [HttpPost(route)] with the route
=>  the response should provide a Response class with the apropriate response and Created when its created
  => should also provide the Get Request that can be used to obtain the created object

Using Mapping
=> Common in APIs and ASPNet
=> There are multiple libraries but mapping should be easy
=> Create a Folder Mapping and a ContractMapping class to represent the extensions that perform the mapping
=> Way better coding => avoid using automapper
=> Do not put business logic in mappers

Keeping track of endpoints
=> Make a central location usually named ApiEndpoints
=> create a private const to represent the base and create partial classes for each controller
=> this allows for better mapping of endpoints
=> Avoids out of sync problems
